```python
import pandas as pd
import numpy as np

np.random.seed(100)
n=500

cities=["ahmedabad","Rajkot","Surat","jamnagar","bhavanagar"]
items=["pizza","burgur","sandwich","pasta"]
payment=["UPI","cash","card"]
status=["delivered","cancelled"]

df=pd.DataFrame({
    "Order_ID": ["fd"+str(1001+i)for i in range(n)],
    "Customer_ID": ["fd_cust"+str(i+1).zfill(3)for i in range(n)],
    "City" : np.random.choice(cities,n),
    "Food_items":np.random.choice(items,n),
    "Quantity":np.random.randint(1,5,n),
    "price":np.random.randint(100,601,n),
    "Delivery_Time":np.random.randint(15,46,n),
    "payament_mode":np.random.choice(payment,n),
    "status":np.random.choice(status,n),
    "Rating":np.random.randint(1,6,n)
})



df.to_csv("Test.csv", index=False)

```
