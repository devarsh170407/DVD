```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(100)
n = 500

cities = ["Ahmedabad","Surat","Rajkot","Vadodara","Jamnagar","Anand"]
account_types = ["Savings", "Current", "Salary", "FD"]
gender = ["Male", "Female"]
loan_status = ["Yes", "No"]

dates = pd.date_range("2023-01-01", "2025-12-31")

df = pd.DataFrame({
    "Customer_ID": np.arange(2001, 2001+n),
    "Date": np.random.choice(dates, n),
    "Age": np.random.randint(22, 60, n),
    "Gender": np.random.choice(gender, n),
    "City": np.random.choice(cities, n),
    "Account_Type": np.random.choice(account_types, n),
    "Balance": np.random.randint(50000, 700000, n),
    "Income": np.random.randint(30000, 180000, n),
    "Loan_Status": np.random.choice(loan_status, n),
    "Credit_Score": np.random.randint(600, 900, n),
    "Transactions": np.random.randint(10, 100, n),
    "Satisfaction": np.random.randint(1, 6, n)
})
```

```python
df.head()
```

```python
df.to_csv("bank_data_new.csv", index=False)
```

```python
plt.figure(figsize=(18,10))
plt.suptitle("Banking Customer Analytics Dashboard", fontsize=18, fontweight='bold')

# 1 Gender
plt.subplot(2,3,1)
sns.countplot(data=df, x="Gender", hue="Gender", palette="pastel", legend=False)
plt.title("Gender Distribution")

# 2 Account Type
plt.subplot(2,3,2)
sns.countplot(data=df, y="Account_Type", hue="Account_Type", palette="coolwarm", legend=False)
plt.title("Account Types")

# 3 City
plt.subplot(2,3,3)
df['City'].value_counts().plot(kind='bar', color='orange')
plt.xticks(rotation=30)
plt.title("City Distribution")

# 4 Income vs Balance
plt.subplot(2,3,4)
sns.scatterplot(data=df, x="Income", y="Balance", hue="Gender")
plt.title("Income vs Balance")

# 5 Credit Score
plt.subplot(2,3,5)
sns.histplot(df["Credit_Score"], bins=20, kde=True, color='green')
plt.title("Credit Score Analysis")

# 6 Satisfaction
plt.subplot(2,3,6)
sns.countplot(data=df, x="Satisfaction", hue="Satisfaction", palette="viridis", legend=False)
plt.title("Customer Satisfaction")

plt.tight_layout()
plt.show()
```
