```python
import pandas as pd

df = pd.read_csv("bank_data_new.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

df['City'].value_counts().plot(
    kind='bar',
    color='orange',
    edgecolor='black'
)

plt.title("Customer Distribution by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.grid(axis='y')

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Age'], bins=15, color='green', edgecolor='black')

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Income'], df['Balance'], color='blue', alpha=0.6)

plt.title("Income vs Balance")
plt.xlabel("Income")
plt.ylabel("Balance")
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['Account_Type'].value_counts().plot(
    kind='bar',
    color='purple',
    edgecolor='black'
)

plt.title("Account Type Distribution")
plt.xlabel("Account Type")
plt.ylabel("Count")
plt.grid(axis='y')

plt.show()
```

```python
plt.figure(figsize=(6,6))

df['Loan_Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['lightgreen','red'],
    shadow=True,
    startangle=90
)

plt.title("Loan Status Distribution")
plt.ylabel("")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Credit_Score'], bins=20, color='teal', edgecolor='black')

plt.title("Credit Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.grid()

plt.show()
```

```python
loan_df = df[df['Loan_Status'] == "Yes"]

plt.figure(figsize=(8,5))

plt.scatter(loan_df['Credit_Score'], loan_df['Balance'], color='red', alpha=0.6)

plt.title("Credit Score vs Balance (Loan Customers)")
plt.xlabel("Credit Score")
plt.ylabel("Balance")
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['Satisfaction'].value_counts().sort_index().plot(
    kind='bar',
    color='gold',
    edgecolor='black'
)

plt.title("Customer Satisfaction")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.grid(axis='y')

plt.show()
```

```python
avg_trans = df.groupby('City')['Transactions'].mean()

plt.figure(figsize=(8,5))

avg_trans.plot(kind='line', marker='o', color='blue')

plt.title("Average Transactions per City")
plt.xlabel("City")
plt.ylabel("Transactions")
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

df.boxplot(column='Balance', by='Account_Type')

plt.title("Balance Distribution by Account Type")
plt.suptitle("")
plt.xlabel("Account Type")
plt.ylabel("Balance")

plt.show()
```

```python
plt.figure(figsize=(8,5))

scatter = plt.scatter(
    df['Age'],
    df['Transactions'],
    c=df['Credit_Score'],
    cmap='viridis'
)

plt.colorbar(scatter)

plt.title("Age vs Transactions")
plt.xlabel("Age")
plt.ylabel("Transactions")
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

# 1 Histogram – Age
plt.subplot(2,2,1)
plt.hist(df['Age'], bins=15, color='green', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

# 2 Bar – Account Type
plt.subplot(2,2,2)
df['Account_Type'].value_counts().plot(
    kind='bar',
    color='purple',
    edgecolor='black'
)
plt.title("Account Types")
plt.xlabel("Type")
plt.ylabel("Count")

# 3 Scatter – Income vs Balance
plt.subplot(2,2,3)
plt.scatter(df['Income'], df['Balance'], color='blue', alpha=0.6)
plt.title("Income vs Balance")
plt.xlabel("Income")
plt.ylabel("Balance")

# 4 Pie – Loan Status (instead of Digital Banking)
plt.subplot(2,2,4)
df['Loan_Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['lightgreen','red'],
    startangle=90
)
plt.title("Loan Status")
plt.ylabel("")

plt.tight_layout()
plt.show()
```
