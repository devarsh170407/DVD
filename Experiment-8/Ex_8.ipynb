```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500 

df = pd.DataFrame({
    "Enrollment_No": np.arange(1, n+1),
    "Student_Name": ["Student"+str(i) for i in range(n)],
    "Surname": ["Surname"+str(i) for i in range(n)],
    "Gender": np.random.choice(["Male", "Female"], n),
    "Department": np.random.choice(["ICT", "CSE", "IT", "AI"], n),
    "Class": np.random.choice(["4EK1", "4EK2", "4EK3"], n),
    "Semester": [8]*n,

    "SGPA_Sem1": np.random.uniform(6,10,n),
    "SGPA_Sem2": np.random.uniform(6,10,n),
    "SGPA_Sem3": np.random.uniform(6,10,n),
    "SGPA_Sem4": np.random.uniform(6,10,n),
    "SGPA_Sem5": np.random.uniform(6,10,n),
    "SGPA_Sem6": np.random.uniform(6,10,n),
    "SGPA_Sem7": np.random.uniform(6,10,n),
    "CGPA": np.random.uniform(6,10,n),

    "HackerRank_Score": np.random.randint(50, 100, n),
    "LeetCode_Problems_Solved": np.random.randint(50, 300, n),
    "GitHub_Repositories": np.random.randint(1, 20, n),
    "Projects_Completed": np.random.randint(1, 10, n),
    "Internship_Completed": np.random.choice(["Yes", "No"], n),

    "Primary_Skill": np.random.choice(["Python", "Java", "Web Dev", "Data Science", "AI"], n),
    "Secondary_Skill": np.random.choice(["Cloud", "DevOps", "ML", "UI UX", "Database"], n),
    "Preferred_Domain": np.random.choice(["Data Science", "Software Dev", "AI", "Web Dev", "Cybersecurity"], n),

    "Preferred_Location": np.random.choice(["Bangalore", "Pune", "Ahmedabad", "Hyderabad", "Remote"], n),
    "Expected_Salary_LPA": np.random.uniform(3, 15, n),

    "Communication_Skill_Rating": np.random.randint(1, 6, n)
})

# Save file
df.to_csv("student_placement_dataset.csv", index=False)

print("Dataset created with shape:", df.shape)
df.head()
```

Question 1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x="Department",
    hue="Department",
    palette="Set2",
    edgecolor="black",
    legend=False           
)
plt.title("Student Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Question 2

```python
plt.figure(figsize=(8,5))
sns.histplot(df["CGPA"], bins=15, kde=True, color="skyblue", edgecolor="black")

plt.title("CGPA Distribution of Students")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.grid(linestyle='--', alpha=0.7)

plt.show()
```

Question 3

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="HackerRank_Score",
    y="CGPA",
    data=df,
    color="purple",
    s=80,
    alpha=0.6
)

plt.title("HackerRank Score vs CGPA")
plt.xlabel("HackerRank Score")
plt.ylabel("CGPA")
plt.grid(linestyle='--', alpha=0.7)

plt.show()
```

Question 4

```python
plt.figure(figsize=(9,5))
sns.countplot(
    data=df,
    x="Preferred_Domain",
    hue="Preferred_Domain",   
    palette="pastel",
    edgecolor="black",
    legend=False
)

plt.title("Preferred Domain Distribution")
plt.xlabel("Domain")
plt.ylabel("Number of Students")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Question 5

```python
internship_counts = df["Internship_Completed"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    internship_counts,
    labels=internship_counts.index,
    autopct='%1.1f%%',
    colors=["lightgreen", "salmon"],
    explode=[0.05, 0],
    shadow=True,
    startangle=90
)

plt.title("Internship Completion Analysis")

plt.show()
```

Question 9

```python

avg_cgpa = df.groupby("Department")["CGPA"].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(
    data=avg_cgpa,
    x="Department",
    y="CGPA",
    marker="o",
    linewidth=2.5,
    color="blue"
)

plt.title("Average CGPA by Department")
plt.xlabel("Department")
plt.ylabel("Average CGPA")
plt.grid(linestyle='--', alpha=0.7)

plt.show()
```

Question 12

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

corr = df[[
    "CGPA",
    "HackerRank_Score",
    "LeetCode_Problems_Solved",
    "GitHub_Repositories",
    "Projects_Completed",
    "Expected_Salary_LPA"
]].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()
```
