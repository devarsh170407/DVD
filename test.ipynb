```python
import pandas as pd
import numpy as np

np.random.seed(100)
n = 500

cities = ["Ahmedabad","Surat","Rajkot","Vadodara","Gandhinagar","Bhavnagar","Jamnagar","Junagadh"]
areas = ["Urban", "Semi-Urban"]
customer_types = ["New", "Regular", "Chronic"]
products = ["Paracetamol","Vitamin Tablets","Glucose Meter","BP Monitor","Sanitizer","Face Wash"]

# Correct mapping (IMPORTANT)
product_data = {
    "Paracetamol": "Medicine",
    "Vitamin Tablets": "Supplement",
    "Glucose Meter": "Device",
    "BP Monitor": "Device",
    "Sanitizer": "Personal Care",
    "Face Wash": "Personal Care"
}

payment_methods = ["UPI","Card","Cash","Wallet"]
status = ["Delivered","Cancelled","Returned"]

dates = pd.date_range("2024-01-01","2025-12-31")

# Step 1: Create DataFrame (without price first)
df = pd.DataFrame({
    "Order_ID": ["med"+str(1001+i) for i in range(n)],
    "Customer_ID": ["med_cst"+str(i+1).zfill(3) for i in range(n)],
    "Date": np.random.choice(dates, n),
    "City": np.random.choice(cities, n),
    "Area": np.random.choice(areas, n),
    "Customer_Type": np.random.choice(customer_types, n),
    "Product_ID": np.arange(9001, 9001+n),
    "Product_Name": np.random.choice(list(product_data.keys()), n),
    "Quantity": np.random.randint(1, 6, n),
    "Delivery_Time(min)": np.random.randint(20, 61, n),
    "Status": np.random.choice(status, n),
    "Payment_Method": np.random.choice(payment_methods, n),
    "Rating": np.random.randint(1, 6, n)
})

# Step 2: Assign Category correctly
df["Category"] = df["Product_Name"].map(product_data)

# Step 3: Assign price based on category
prices = []

for cat in df["Category"]:
    if cat in ["Medicine", "Supplement"]:
        prices.append(np.random.randint(100, 801))
    elif cat == "Device":
        prices.append(np.random.randint(500, 3001))
    else:  # Personal Care
        prices.append(np.random.randint(50, 401))

df["Selling_Price"] = prices

# Step 4: Cost price (65%–85%)
df["Cost_Price"] = (df["Selling_Price"] * np.random.uniform(0.65, 0.85, n)).astype(int)

# Save CSV
df.to_csv("medical_orders.csv", index=False)

print("CSV generated successfully!")
```
