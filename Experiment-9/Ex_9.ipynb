```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

avg_hours = df.groupby("Course_Difficulty")["Hours_Spent_Per_Week"].mean().reset_index()

fig = px.line(
    avg_hours,
    x="Course_Difficulty",
    y="Hours_Spent_Per_Week",
    title="Study Hours by Course Difficulty"
)

fig.update_traces(
    mode="lines+markers",
    marker=dict(size=10, color="red")
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

category_count = df["Course_Category"].value_counts().reset_index()
category_count.columns = ["Course_Category", "Count"]


fig = px.bar(
    category_count,
    x="Course_Category",
    y="Count",
    title="Students Enrolled in Each Course Category",
    labels={"Course_Category": "Course Category", "Count": "Number of Students"}
)


fig.update_traces(
    width=0.5,
    marker_color="skyblue",
    marker_line_color="black",
    marker_line_width=2,
    text=category_count["Count"]   
)


fig.update_xaxes(
    tickangle=45,
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_yaxes(
    tickfont=dict(size=14),
    title_font=dict(size=16)
)


fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

completion_counts = df["Completion_Status"].value_counts().reset_index()
completion_counts.columns = ["Completion_Status", "Count"]

fig = px.pie(
    completion_counts,
    names="Completion_Status",
    values="Count",
    title="Course Completion Distribution"
)

fig.update_traces(
    textinfo="percent+label",
    hoverinfo="label+percent+value"
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("online_learning_platform_dataset.csv")

fig = px.box(
    df,
    x="Course_Category",
    y="Quiz_Score",
    title="Quiz Performance by Course Category"
)

fig.update_traces(
    marker_color="royalblue"
)

fig.update_xaxes(
    tickangle=45,
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_yaxes(
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

fig = px.violin(
    df,
    x="Course_Difficulty",
    y="Hours_Spent_Per_Week",
    title="Study Hours Comparison by Course Difficulty"
)

fig.update_traces(
    box_visible=True,
    meanline_visible=True
)

fig.update_xaxes(
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_yaxes(
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

stages = ["Students Enrolled", "Videos Watched", "Assignments Submitted", "Quiz Attempted", "Course Completed"]

values = [
    len(df),
    df["Videos_Watched"].sum(),
    df["Assignments_Submitted"].sum(),
    len(df[df["Quiz_Score"] > 0]),
    len(df[df["Completion_Status"] == "Completed"])
]

fig = px.funnel(
    x=values,
    y=stages,
    title="Learning Engagement Funnel"
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("online_learning_platform_dataset.csv")

fig = px.scatter(
    df,
    x="Assignments_Submitted",
    y="Quiz_Score",
    title="Assignments Submitted vs Quiz Score"
)

fig.update_traces(
    marker=dict(
        size=10,
        color="royalblue",
        opacity=0.7
    )
)

fig.update_xaxes(
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_yaxes(
    tickfont=dict(size=14),
    title_font=dict(size=16)
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

city_category = df.groupby(["City", "Course_Category"]).size().reset_index(name="Count")

fig = px.treemap(
    city_category,
    path=["City", "Course_Category"],
    values="Count",
    title="Student Distribution by City and Course Category"
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

sunburst_data = df.groupby(
    ["Course_Category", "Course_Difficulty", "Completion_Status"]
).size().reset_index(name="Count")

fig = px.sunburst(
    sunburst_data,
    path=["Course_Category", "Course_Difficulty", "Completion_Status"],
    values="Count",
    title="Learning Activity Hierarchy"
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```

```python
df = pd.read_csv("online_learning_platform_dataset.csv")

corr = df[[
    "Hours_Spent_Per_Week",
    "Videos_Watched",
    "Assignments_Submitted",
    "Quiz_Score"
]].corr()

fig = px.imshow(
    corr,
    text_auto=True,
    title="Learning Metrics Correlation"
)

fig.update_layout(
    title_font=dict(size=20)
)

fig.show()
```
